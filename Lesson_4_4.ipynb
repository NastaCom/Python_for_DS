```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
credcard = pd.read_csv('C:/Users/User/creditcard.csv')
credcard
```

```python
Class = credcard.value_counts()
Class
```

```python
Class.plot(kind='hist')
plt.show()
```

```python
Class.plot(kind='hist', logx=True)
plt.show()
```

```python
class0 = credcard.loc[credcard['Class'] == 0, ['V1']]
class1 = credcard.loc[credcard['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.show()
```
