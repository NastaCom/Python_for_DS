С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно. Создайте 
массив Numpy под названием f, содержащий косинусы элементов массива t. Постройте линейную диаграмму, используя 
массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета. 
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для 
вертикальной - 'Значения f'. Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f)
plt.show()
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, label='f(t)', color='green')
plt.show()
```

```python
title_dict = {'fontsize': 18, 'color': 'grey', 'family': 'Calibri'}
```

```python
plt.plot(t, f)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, label='f(t)', color='green')
plt.title('График f(t)', fontdict=title_dict)
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```
